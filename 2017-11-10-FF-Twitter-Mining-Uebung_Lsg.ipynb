# Aufgabe 2

Suche nach Tweets, die @donaldtrump (dies ist nicht der richtige Account von Donald Trump 😉 ) erwähnen. Sammle die ersten 100 Ergebnisse von deiner Suche. Und lass dir die ersten 5 anzeigen. (Hinweis: https://developer.twitter.com/en/docs/tweets/search/api-reference/get-search-tweets) 



```python
q = '@donaldtrump' 
count = 100

search_donald = twitter_api.search.tweets(q=q, count=count)

statuses = search_donald['statuses']

print(json.dumps(statuses[:5], indent=1))
```

Gib die ersten 5 Tweets, Usernamen und Hashtags aus deiner Sammlung aus.



```python
status_texts = [ status['text'] for status in statuses ] #Tweets
    
usernames = [ user_mention['screen_name'] for status in statuses 
                                        for user_mention in status['entities']['user_mentions'] ] #User

hashtags = [ hashtag['text'] for status in statuses
                                       for hashtag in status['entities']['hashtags']] #Hashtags

print(json.dumps(status_texts[0:5], indent=1))
print(json.dumps(usernames[0:5], indent=1))
print(json.dumps(hashtags[0:5], indent=1))
```

Erstelle eine Sammlung aller Wörter, die in den Tweets vorkommen und gib die ersten 30 Wörter aus.

```python
words = [ w 
             for statuses in status_texts
                for w in statuses.split() ]

print(json.dumps(words[:30], indent=1))
```

Erstelle nun eine Häufigkeitsverteilung der Elemente *Wörter*, *Usernamen* und *Hashtags*. Lass dir die Plätze 5-15 in einer Tabelle anzeigen.

```python
from collections import Counter
from prettytable import PrettyTable


for label, data in (('Word', words),
                    ('Screen Name', usernames),
                    ('Hashtag', hashtags)):
    pt = PrettyTable(field_names=[label, 'Count'])
    
    c = Counter(data)
    [ pt.add_row(kv) for kv in c.most_common()[4:15] ]
    pt.align[label], pt.align['Count'] = 'l', 'r' # Spaltenausrichtung festlegen
    print(pt)
```

Gib nun die 5 am meisten geteilten Tweets in einer Tabelle aus.

```python
retweets = [
            # Speichern eines Tupels dieser 3 Werte...
            (status['retweet_count'],
             status['retweeted_status']['user']['screen_name'],
             status['text'])
            
            #... für jeden Status...
            for status in statuses
    
            #... sofern der Status diese Bedingung erfüllt
                if 'retweeted_status' in status
            ]

#Ausgabe der ersten 5 des sortierten Ergebnisses
pt = PrettyTable(field_names=['Count', 'Screen Name', 'Text'])
#Darstellung jedes Elemens im Tupel
[ pt.add_row(row) for row in sorted(retweets, reverse=True)[:5] ] 
pt.max_width['Text'] = 50
pt.align = 'l'
print(pt)
```
